```python
from matplotlib import rc
import numpy as np
import matplotlib.pyplot as plt

R = 1000 #Ohms
C = 1e-6 #Farads
V = 5 #Volts

t = np.linspace(0, 0.01, 1000)
Vc = V * (1 - np.exp(-t / (R * C)))

plt.plot(t, Vc)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('RC Charging Curve')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


R = 1000 # Ohms
C = 1e-6 # Farads
V = 5 # Volts

#Time array
t = np.linspace(0, 0.01, 1000)
Vc_charge = V * (1 - np.exp(-t / (R * C)))
#Modified discharging equation
Vc_discharge = V * np.exp(-t / (R * C))
#Plot both on the same graph
plt.plot(t, Vc_charge, label= 'Charging')
plt.plot(t, Vc_discharge, label= 'Discharging')


plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("RC Circuit Charging and Discharging Curve")
plt.legend()
plt.grid(True)
plt.show()
```
