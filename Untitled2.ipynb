```python
import random, time
import matplotlib.pyplot as plt

temps = []
for t in range(10):
  temp = 25 + random.uniform(-1, 1)
  temps.append(temp)
  with open("temps.txt", "a") as f:
    f.write(str(temp) + "\n")
    time.sleep(0.5)
plt.plot(temps)
plt.xlabel("Time (s)")
plt.ylabel("Temperature (C)")
plt.title("Temperature Log")
plt.show()
```

```python
import csv
import random
import time
import matplotlib.pyplot as plt


# Data storage lists
time_data = []
temperature_data = []
humidity_data = []

try:
    with open("environment_data.csv", "w", newline="") as csvfile, open("environment_data.txt", "w") as txtfile:
          writer = csv.writer(csvfile)
          writer.writerow(["Time", "Temperature", "Humidity"])


          for t in range(10):
              temp = random.uniform(20, 30)
              hum = random.uniform(40, 70)


              time_data.append(t)
              temperature_data.append(temp)
              humidity_data.append(hum)


              writer.writerow([t, temp, hum])
              txtfile.write(f"{t}\t{temp:.2f}\t{hum:.2f}\n")
              time.sleep(0.5)

except IOError as e:
    print("File error:", e)


plt.plot(time_data, temperature_data, label="Temperature")
plt.plot(time_data, humidity_data, label="Humidity")
plt.xlabel("Time (s)")
plt.ylabel("Values")
plt.title("Environmental Data vs Time")
plt.legend()
plt.show()
```
